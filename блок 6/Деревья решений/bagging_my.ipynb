```python
import pandas as pd
import numpy as np

from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import RandomForestRegressor

from sklearn import metrics

from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
```

```python
data = pd.read_csv('data/wineQualityReds.csv')
data
```

```python
data['quality'].describe()
```

```python
data['label'] = data['quality'].apply(lambda x:1 if x>=6 else 0)
```

```python
X = data.drop(['quality','label'], axis = 1)
y = data.label

X.shape, y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=42)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
lr = LogisticRegression()
lr.fit(X_train, y_train)

y_lr_test = lr.predict(X_test)

print('Значение f1 для логистической регрессии {:.3f}'.format(metrics.f1_score(y_test, y_lr_test)))
```

```python
dt = DecisionTreeClassifier(random_state=42, max_depth=10)
dt.fit(X_train, y_train)

y_dt_test = dt.predict(X_test)

print('Значение f1 для дерева решений {:.3f}'.format(metrics.f1_score(y_test, y_dt_test)))
```

```python
bag_cl = BaggingClassifier(DecisionTreeClassifier(random_state=42, max_depth=10), n_estimators=1500, random_state=42)

bag_cl.fit(X_train, y_train)

y_bag_test = bag_cl.predict(X_test)

print('Значение f1 для бэггинга {:.3f}'.format(metrics.f1_score(y_test, y_bag_test)))
```

```python
model = BaggingClassifier(base_estimator=DecisionTreeClassifier(max_depth = 10, random_state=42), n_estimators=1500, random_state=42).fit(X_train, y_train)
preds_train = model.predict(X_train)
preds_test = model.predict(X_test)
metrics.f1_score(preds_test, y_test)
```

```python
boston = pd.read_csv('data/boston.csv', decimal='.')
boston
```

```python
columns = list(boston.columns)
col_drop = ['river', 'highways_index', 'tax']
for col_dr in col_drop:
    columns.remove(col_dr)
    
columns
```

```python
for col in columns:
    boston[col] = boston[col].apply(lambda x: x.replace(',','.'))
    boston[col] = boston[col].astype(float)
```

```python
boston.info()
```

```python
X = boston.drop('target', axis = 1)
y = boston.target

X.shape, y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 13, train_size=0.7)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
y_train.median()
```

```python
lr = LinearRegression()

lr.fit(X_train, y_train)
y_pred_test = lr.predict(X_test)

round(metrics.mean_absolute_error(y_test, y_pred_test), 2)
```

```python
dr = DecisionTreeRegressor(random_state = 13)

dr.fit(X_train, y_train)
y_pred_train = dr.predict(X_train)
y_pred_test = dr.predict(X_test)

print(round(metrics.mean_absolute_error(y_train, y_pred_train), 2))
print(round(metrics.mean_absolute_error(y_test, y_pred_test), 2))
```

```python
rfr1 = RandomForestRegressor(n_estimators=3, random_state = 13)

rfr1.fit(X_train, y_train)
y_pred_train = rfr1.predict(X_train)
y_pred_test = rfr1.predict(X_test)

print(round(metrics.mean_absolute_error(y_train, y_pred_train), 2))
print(round(metrics.mean_absolute_error(y_test, y_pred_test), 2))
```

```python
df = pd.read_csv('data/weatherAUS.csv')
df
```

```python
df.info()
```

```python
df.isnull().sum().sum()
```

```python
(df.isnull().sum()/df.shape[0] * 100).sort_values()
```

```python
df.drop(['Cloud3pm', 'Evaporation', 'Sunshine'], axis=1, inplace=True)
df.info()
```

```python
df['RainToday'] = df['RainToday'].apply(lambda x: 1 if x=='Yes' else 0)
df['RainTomorrow'] = df['RainTomorrow'].apply(lambda x: 1 if x=='Yes' else 0)
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Date']
```

```python
df['Month'] = df['Date'].dt.month
df.drop('Date', axis=1, inplace=True)
```

```python
df.groupby('Month')['RainToday'].mean().sort_values()
```

```python
df.info()
```

```python
categoricals = ['Month','Location', 'WindGustDir', 'WindDir9am', 'WindDir3pm']
df_dummies = pd.get_dummies(df, columns=categoricals)
df_dummies
```

```python
df_dummies.dropna(inplace=True)
```

```python
X = df_dummies.drop('RainTomorrow', axis=1)
y = df_dummies['RainTomorrow']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=31)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
y_test.mean()
```

```python
inds = np.random.randint(0, len(X_train), len(X_train))
inh = X_train.iloc[inds,:]
inh
```

```python
def get_random_ids(data):
    inds = np.random.randint(0, len(data), len(data))
    data = data[inds]
    mean_dt = data.mean()
    return mean_dt
```

```python
target = X_train['MinTemp'].values #выбираем целевую переменную
#np.random.seed(31)
mean_values = [get_random_ids(target) for i in range(1000)]
np.std(mean_values)
```

#### Логистическая регрессия

```python
lr = LogisticRegression()

lr.fit(X_train, y_train)

y_pred_train = lr.predict(X_train)
y_pred_test = lr.predict(X_test)

round(metrics.roc_auc_score(y_test, y_pred_test), 2)
```

#### GridSearchCV 

```python
params = {'max_leaf_nodes': list(range(5, 15)), 'min_samples_split': [2, 3, 4], 'max_depth': [5,7,9,11]}

gsv = GridSearchCV(
    estimator=DecisionTreeClassifier(random_state=42), 
    param_grid = params,
    cv = 3,
    return_train_score = True
    )

gsv.fit(X_train, y_train)
```

```python
gsv.best_params_
```

```python
dt = DecisionTreeClassifier(random_state=42, max_depth=5, max_leaf_nodes= 12, min_samples_split=2)

dt.fit(X_train, y_train)

y_pred_train = dt.predict(X_train)
y_pred_test = dt.predict(X_test)

round(metrics.roc_auc_score(y_test, y_pred_test), 2)
```

#### Случайный лес

```python
rf = RandomForestRegressor(n_estimators=100, random_state=31)

rf.fit(X_train, y_train)

y_pred_train = rf.predict(X_train)
y_pred_test = rf.predict(X_test)

round(metrics.roc_auc_score(y_test, y_pred_test), 2)
```

#### GridSearchCV и случайный лес

```python
params = {'max_features': [ 4, 5, 6, 7], 'min_samples_leaf': [3, 5, 7, 9, 11], 'max_depth': [5, 10, 15]}

gsv = GridSearchCV(
    estimator=RandomForestRegressor(n_estimators=100, random_state=31), 
    param_grid = params,
    cv = 3,
    return_train_score = True
    )

gsv.fit(X_train, y_train)
```

```python
gsv.best_params_
```

```python
rf = RandomForestRegressor(n_estimators=100, random_state=31, max_depth = 15, max_features = 7, min_samples_leaf = 3)

rf.fit(X_train, y_train)

y_pred_train = rf.predict(X_train)
y_pred_test = rf.predict(X_test)

round(metrics.roc_auc_score(y_test, y_pred_test), 2)
```
